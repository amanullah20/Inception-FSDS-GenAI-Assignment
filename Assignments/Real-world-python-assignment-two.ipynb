1. Email Validator (String Filtering + Loops) 

```python
def is_valid_email(email):
    if "@" in email and email.endswith('.com') or email.endswith('.org'):
        print("Valid email")

    else:

        print("invalid email")    
```

2. Count Word Frequency in Customer Reviews 

```python
def word_count(text):
    text=text.lower()
    words = text.split()
    count = {}

    for word in words:
        if word in count:
            count[word]+=1 
        else:
            count[word]=1 

    return count 

review="This product is good and this is excellent Product"
result = word_count(review)
print(result)    
```

 # 3.Password Strength Checker

```python
def is_strong_password(password):
    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False
    
    special_characters = "!@#$%^&*()-_=+[]{}|;:',.<>?/"

    for char in password:
        if char.isupper():
            has_upper = True
        elif char.islower():
            has_lower = True
        elif char.isdigit():
            has_digit = True
        elif char in special_characters:
            has_special = True

    if has_upper and has_lower and has_digit and has_special:
        return True
    else:
        return False


# Example usage
password = "Hello@123"

if is_strong_password(password):
    print("Strong Password")
else:
    print("Weak Password")
```

# 4. Username Generator

```python
import random

def generate_username(full_name):
    # Spliting first and last name
    names = full_name.lower().split()
    first = names[0]
    last = names[1]

    first_part = ""
    last_part = ""

    # Geting first 3 letters of first name (using loop)
    for i in range(len(first)):
        if i < 3:
            first_part += first[i]

    # Geting last 3 letters of last name (using loop)
    for i in range(len(last)):
        if i >= len(last) - 3:
            last_part += last[i]

    # Generating a random 2-digit number
    random_number = random.randint(10, 99)

    # Final username
    username = first_part + last_part + str(random_number)
    return username


```

```python
# Example usage
name = "Mamun Malitha"
print(generate_username(name))
```

# 5. SMS Word Counter 

```python
def sms_checker(message):
    # Count characters
    char_count = len(message)

    # Count words
    words = message.split()
    word_count = len(words)

    # Check SMS limit
    if char_count > 160:
        status = "Exceeded Limit"
    else:
        status = "Within Limit"

    return word_count, char_count, status


```

```python
# Example usage
msg = "This is a demo message."
words, chars, status = sms_checker(msg)

print("Words =", words)
print("Characters =", chars)
print("Status =", status)
```

# 6. Remove Duplicates from Contact List

```python
def remove_duplicates(name_string):
    # Split names and remove extra spaces
    names = [name.strip() for name in name_string.split(",")]

    unique_names = []
    
    # Loop to remove duplicates
    for name in names:
        if name not in unique_names:
            unique_names.append(name)

    # Join back into comma-separated string
    return ", ".join(unique_names)



```

```python

input_names = "Rahim, Karim, Rahim, Sultana"
output = remove_duplicates(input_names)
print(output)
```

# 7. Detect Suspicious Words in Text (Moderation)

```python
def detect_banned_words(text):
    banned_words = ['spam', 'scam', 'fake']

    # Convert text to lowercase
    text_lower = text.lower()

    # Check if any banned word exists
    for word in banned_words:
        if word in text_lower:
            return "Warning: Suspicious content detected"
    
    return "Content OK"



```

```python

msg = "This is a scam offer"
print(detect_banned_words(msg))
```

# 8. Invoice Code Formatter 

```python
def format_invoice(number):
    # Convert number to string
    num_str = str(number)

    # Add leading zeros to make it 5 digits
    while len(num_str) < 5:
        num_str = "0" + num_str

    # Add prefix
    return "INV" + num_str


```

```python

invoice = 23
print(format_invoice(invoice))
```

# 9. Count Hashtags and Mentions (Social Media Analytics) 

```python
def count_tags(post):
    hashtag_count = 0
    mention_count = 0

    for char in post:
        if char == '#':
            hashtag_count += 1
        elif char == '@':
            mention_count += 1

    return hashtag_count, mention_count


```

```python

text = "Check out #AIWorkshop and follow @inceptionbd"
hashtags, mentions = count_tags(text)

print("Hashtags =", hashtags)
print("Mentions =", mentions)
```

# 10. Word Reversal for Data Encryption

```python
def reverse_each_word(sentence):
    words = sentence.split()
    reversed_words = []

    for word in words:
        reversed_word = ""
        for char in word:
            reversed_word = char + reversed_word
        reversed_words.append(reversed_word)

    return " ".join(reversed_words)


```

```python

text = "Aman ullah"
print(reverse_each_word(text))
```
