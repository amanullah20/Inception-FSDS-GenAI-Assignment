# 1. Inventory Price Calculator

```python
def calculate_inventory_summary(price):
    total_price=0
    for i in price:
        total_price += i 

    average = total_price/len(price)    
    return total_price , average     
```

```python
listOfPrices = [120, 250, 399, 150] 
result= calculate_inventory_summary(listOfPrices)
print(result)
```

# 2. Unique Visitor Tracker

```python
def count_unique_visitors(visitor_list):
    unique_visitor = []

    for num in visitor_list:
        if num not in unique_visitor:
            unique_visitor.append(num) 

    return unique_visitor
```

```python
visitors = [101, 205, 101, 310, 205, 550]
unique_ids = count_unique_visitors(visitors)

print(f"total unique visitors {len(unique_ids)} and they are {unique_ids}" )
```

# 3. Product Category Counter

```python
def count_categories(products):
    category_cout = {}

    for itm in products:
        if itm in category_cout:
            category_cout[itm] += 1

        else:
            category_cout[itm] = 1 

    return category_cout 
```

# 4. Temperature Analytics

```python
def analyze_temperatures(temp_list):
    high_list = []
    normal_list = [] 

    for temp in temp_list:
        if temp <= 38:
            normal_list.append(temp) 
        else:
            high_list.append(temp) 
    return high_list,normal_list 

            
```

```python
temp_list = [36.5, 38.0, 39.1, 37.2, 40.5, 38.0, 37.9]
high_temp_list , normal_tmp_list = analyze_temperatures(temp_list)
```

```python
print(f"High_list (>38c): {high_temp_list} \n Normal List(<38 c) : {normal_tmp_list} \n Full output Tuple : {high_temp_list,normal_tmp_list} "  )
```

# 5. Menu Price Lookup

```python
def get_price(menu, item):
    if item in menu:
        return menu[item]
    else:
        return "item not found"
```

```python
menu = {"Coffee": 3.50, "Tea": 2.75, "Muffin": 4.00, "Sandwich": 8.50}
print(get_price(menu, "Muffin"))
```

```python
print(get_price(menu, "Donut"))
```

# 6. Student Grading System

```python
def calculate_grade(score):
    if score >=80:
        return "A+"
    elif score >= 70:
        return "A" 
    elif score >= 60:
        return "A-" 
    else:
        return "F" 
```

```python
def students_grade(std_dict):
    grd_dict = {}

    for student , score in std_dict.items():
        grd_dict[student] = calculate_grade(score)  

    return grd_dict

```

```python
scores = {"Alice": 85, "Bob": 78, "Charlie": 32, "David": 69}
grades = students_grade(scores)
print(grades)
```

# 7. Discount Calculator

```python
def apply_discou(products):
    updated_prices = []

    for name , price, discount in products:
        new_price = price - (price * discount/100)
        updated_prices.append(new_price) 

    return updated_prices
```

```python
items = [
    ("Keyboard", 80.00, 20),
    ("Mouse", 25.00, 10),
    ("Monitor", 300.00, 5)
]

apply_discou(items)
```

# 8. Hashtag Generator

```python
def generate_hashtags(sentence):
    words = sentence.lower().split()

    hashtags = [] 

    for w in words:
        hashtags.append("#" + w)

    return hashtags    
```

```python
text = "AI Data Science"
generate_hashtags(text)
```
