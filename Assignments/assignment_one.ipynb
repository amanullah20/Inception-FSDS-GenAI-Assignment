# Assignment 2: ATM Withdrawal System 

```python
class Atmmachine:
    def __init__(self):
        self.pin = ""
        self.balance=0
        self.menu()

    def menu(self):
        user_input = input(
            """
            How can i help you?
            01 . press 1 to create pin
            02 . press 2 to change pin
            03 . press 3 to check balance
            04 . press 4 to withdraw
            05 . anything to exit
            """
        ) 

        if user_input == "1":
            # create pin
            self.create_pin()
        elif user_input=="2":
            # change pin 
            self.change_pin()
        elif user_input=="3":
            # check balance
            self.check_balance()

        elif user_input=="4":
            # withdraw money 
            self.withdraw_balance()
        else:
           exit()

    def create_pin(self):
       user_pin = input("please enter your pin")
       self.pin = user_pin
       user_balance= int(input("please enter your balance"))
       self.balance=user_balance
       print("your pin is created successfully") 
       self.menu()

    def change_pin(self):
        old_pin = input("enter your old pin:" )
        if old_pin == self.pin:
            new_pin = input("enter your new Pin: ")
            self.pin = new_pin
            print("New pin created successfully")
            self.menu()
        else:
            print("Invalid pin")
            self.menu()
        
    def check_balance(self):
        user_pin = input("enter your pin")
        if user_pin == self.pin:
            print(f"your balance is {self.balance}")
            self.menu() 
        else:
            print("Invalid Password")
            self.menu()
        
    def withdraw_balance(self):
        user_pin = input("enter your pin")
        if user_pin == self.pin:
            amount = int(input("enter withdraw amount"))
            if amount <= self.balance:
                self.balance = self.balance - amount
                print(f"you withdrew {amount} and your present balance is {self.balance}")
                self.menu() 
            else:
                print("Isufficient balance")
                self.menu()
```

```python
Atm=Atmmachine()
```

```python
class BusFare:
    def __init__(self):
        self.fare=800 
        self.FareCount()

    def FareCount(self):
        age = int(input("Please tell your age for discount"))
        if age < 5:
            print("You are free for the journy ")
        elif age >5 and age <=18:
            self.fare = self.fare * .50
            print(f"your fare is {self.fare}")
        elif age >=60 :
            self.fare= self.fare * .70
            print(f"your fare is {self.fare}")
        else:
            print(f"your fare is {self.fare}")
```
